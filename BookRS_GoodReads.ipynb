Processing missing values and remove unneeded columns

```python
df.dropna(inplace=True)  # Delete rows containing missing values
df.drop('img', axis=1, inplace=True) # Unneeded columns
df.drop('link', axis=1, inplace=True) # Unneeded columns
df.drop('isbn', axis=1, inplace=True) # Unneeded columns
df.drop('isbn13', axis=1, inplace=True) # Unneeded columns
df.head(10)
```

Viewing Data Types

```python
print(df.dtypes)
```

Convert Data Types

```python
df['rating'] = df['rating'].astype(float)
df['pages'] = df['pages'].astype(int)
df['reviews'] = df['reviews'].astype(int)
df['totalratings'] = df['totalratings'].astype(int)
df['genre'] = df['genre'].astype('category')
print(df.dtypes)
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
print(df.shape)
```

```python
# Split the training set and the temporary test set
train_df, temp_test_df = train_test_split(df, test_size=0.3, random_state=30)

# Split the temporary test set into a validation set and a test set and 
val_df, test_df = train_test_split(temp_test_df, test_size=0.5, random_state=30)

# Checking the splitting results
print(f"Training set size: {train_df.shape[0]} ({len(train_df)/len(df)*100:.2f}%)")
print(f"Validation set size: {val_df.shape[0]} ({len(val_df)/len(df)*100:.2f}%)")
print(f"Testing set size: {test_df.shape[0]} ({len(test_df)/len(df)*100:.2f}%)")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder, MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Feature Scaling: Normalization and Standardization
# 初始化标准化器
std_scaler = StandardScaler()

# 对'reviews', 和 'totalratings' 应用标准化
df['reviews'] = std_scaler.fit_transform(df[['reviews']])
df['totalratings'] = std_scaler.fit_transform(df[['totalratings']])

# 初始化归一化器
min_max_scaler = MinMaxScaler()

# 对 'rating' , 'pages'应用归一化
df['rating'] = min_max_scaler.fit_transform(df[['rating']])
df['pages'] = min_max_scaler.fit_transform(df[['pages']])

# 接下来，您可以输出变换后的数据以验证
print(df[['pages', 'reviews', 'totalratings', 'rating']].describe())

# 标签编码 'bookformat' 列
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['bookformat_encoded'] = label_encoder.fit_transform(df['bookformat'])

# 使用 MultiLabelBinarizer 分别对 'genre' 和 'author' 列进行编码
mlb_genre = MultiLabelBinarizer()
genre_encoded = mlb_genre.fit_transform(df['genre'])
genre_encoded_df = pd.DataFrame(genre_encoded, columns=mlb_genre.classes_)

mlb_author = MultiLabelBinarizer()
author_encoded = mlb_author.fit_transform(df['author'])
author_encoded_df = pd.DataFrame(author_encoded, columns=mlb_author.classes_)

# 将编码后的列添加到原始 DataFrame 中
df = pd.concat([df, genre_encoded_df, author_encoded_df], axis=1)

# 删除原始的 'genre' 和 'author' 列
df.drop(columns=['genre', 'author'], inplace=True)

# 输出转换后的数据
df.head(10)


# Additional feature processing as needed
```
